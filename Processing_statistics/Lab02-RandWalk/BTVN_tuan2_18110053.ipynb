```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Rand_walk(nstep):
    position = 0
    arr = [position]
    for i in range(nstep):
        if (np.random.randint(0,2,size=1)==1):
            position += 1
        else:
            position -= 1
        arr.append(position)
    return arr
```

```python
x = Rand_walk(1000)
```

```python
plt.plot(x)
plt.show()
```
