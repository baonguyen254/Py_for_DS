```python
#Thư viện
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
X_0 = np.random.rand(1000, 1)
y = 4 + 3 * X_0 + .2*np.random.randn(1000, 1)
y.shape
```

## BGD

```python
X_0.shape[0]
```

```python
#Khởi tạo lr
np.random.seed(1)
lr= 0.0001  
#Khởi tạo W
W=np.asarray([[2], [1]])
#Thêm một cột cho bias
X=np.concatenate([np.ones([X_0.shape[0],1]),X_0],axis=1) 
print("shape của X là:", X.shape)
print()
#GD
for i in range(1000000):#gồm 10 vòng lặp
  prediction=np.matmul(X,W)
  loss=prediction-y
  gradient=np.matmul(X.T,loss)
  W=W-lr*(1/X.shape[0])*gradient
  if i % 1000 == 0:
      loss = np.mean(np.square(np.matmul(X,W)-y))
      print("Loss at iter {}: {}".format(i, loss))
loss=np.mean(np.square(np.matmul(X,W)-y))
print("Final loss: ", loss)
print("Final W: ", W)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_0,y)
plt.plot(X_0,np.matmul(X,W),"r")
```

## SGD

```python
#Khởi tạo lr
lr=0.0001
#Khởi tạo W
W=np.asarray([[2], [1]])
#Thêm một cột cho bias
X=np.concatenate([np.ones([X_0.shape[0],1]),X_0],axis=1) 
print("shape của X là:", X.shape)
print()
for i in range(10000):#gồm 100000 vòng lặp
  rd_id = np.random.permutation(X.shape[0])
  for j in rd_id:
    X_1=X[j,:].reshape(1,-1)
    error=np.matmul(X_1,W)-y[j,:].reshape(1,-1)
    W=W-lr*np.matmul(X_1.T,error)
  if i % 1000 == 0:
      loss = np.mean(np.square(np.matmul(X,W)-y))
      print("Loss at iter {}: {}".format(i, loss))
loss=np.mean(np.square(np.matmul(X,W)-y))
print("Loss: ", loss)
print("W: ", W)
```

## Câu 3

```python

def BGD(X_train,y_train, lr = 0.001):
    '''
    Batch Gradient Descent
    Input:
        X_train: data_train(test)
        y_train: label_train(test)
        lr: learning_rate,  default = 0.001
    Output:
        Weight set 
        loss: Mean Squared Error Loss  
        W: Estimated coefficients for the linear regression problem
    '''
    # khởi tạo W
    W = np.random.normal(loc = 0,scale=1 ,size = X_train.shape[1]+1).reshape(X_train.shape[1]+1,1)
    #Thêm một cột cho bias
    X = np.concatenate([np.ones([X_train.shape[0],1]),X_train], axis = 1)
    #GD
    for i in range(100000):
        prediction = np.matmul(X, W)
        loss = prediction - y_train
        gradient = np.matmul(X.T, loss)
        W = W -lr*(1/X.shape[0])*gradient
        if i % 1000 == 0:
            loss = np.mean(np.square(np.matmul(X,W)-y_train)) # MSE
    loss = np.mean(np.square(np.matmul(X,W)-y_train))
    print("Loss: ", loss)
    print("W: ", W)
    return loss, W

def SGD(X_train,y_train, lr = 0.001):
    '''
    Stochatic Gradient Descent
    Input:
        X_train: data_train(test)
        y_train: label_train(test)
        lr: learning_rate,  default = 0.001
    Output:
        Weight set 
        loss: Mean Squared Error Loss  
        W: Estimated coefficients for the linear regression problem
    '''
    # khởi tạo W
    W = np.random.normal(loc = 0, scale=1 ,size = X_train.shape[1]+1).reshape(X_train.shape[1]+1,1)
    #Thêm một cột cho bias
    X = np.concatenate([np.ones([X_train.shape[0],1]),X_train], axis = 1)
    for i in range(10000):#gồm 100000 vòng lặp
        rd_id = np.random.permutation(X.shape[0])
        for j in rd_id:
            X_1=X[j,:].reshape(1,-1)
            error=np.matmul(X_1,W)-y_train[j,:].reshape(1,-1)
            W=W-lr*np.matmul(X_1.T,error)
        if i % 1000 == 0:
            loss = np.mean(np.square(np.matmul(X,W)-y_train))
    loss=np.mean(np.square(np.matmul(X,W)-y_train))
    print("Loss: ", loss)
    print("W: ", W)
    return loss, W

def Mini_GD(X_train,y_train, lr = 0.001):
    '''
    Mini Batch Gradient Descent
    Input:
        X_train: data_train(test) 
        y_train: label_train(test)
        lr: learning_rate,  default = 0.001
    Output:
        Weight set 
        loss: Mean Squared Error Loss  
        W: Estimated coefficients for the linear regression problem
    '''
    #Khởi tạo W
    W = np.random.normal(loc = 0,scale=1 ,size = X_train.shape[1]+1).reshape(X_train.shape[1]+1,1)
    #Thêm một cột cho bias
    X = np.concatenate([np.ones([X_train.shape[0],1]),X_train], axis = 1)
    first_X=X.copy()
    first_y=y_train.copy()
    for i in range(200000):#gồm 20000 vòng lặp
        indices = np.random.permutation(X.shape[0])
        first_X=first_X[indices]
        first_y=first_y[indices]

        temp=100
        z=0
        for j in range(0,X.shape[0],128):
            z+=1
            X_1=first_X[j:temp,:]
            prediction=np.matmul(X_1,W)
            loss=prediction-first_y[j:temp,:]
            gradient=np.matmul(X_1.T,loss)
            W=W-lr*(1/X_1.shape[0])*gradient

            temp=temp+ 128 if X.shape[0]-temp>=128 else X.shape[0]
        if i % 1000 == 0:
            loss = np.mean(np.square(np.matmul(X,W)-y_train))
    loss=np.mean(np.square(np.matmul(X,W)-y_train))
    print("Loss: ", loss)
    print("W: ", W)
    return loss, W
```

```python

dict_W_Loss = dict()
y_hat = []
print('BGD')
loss, W = BGD(X_train,y_train, lr=0.001)
y_pred = np.dot(X_test,W)
dict_W_Loss[loss] = W
y_hat.append(y_pred)
print()

print('SGD')
loss, W = SGD(X_train,y_train, lr=0.001)
y_pred = np.dot(X_test,W)
dict_W_Loss[loss] = W
y_hat.append(y_pred)
print()

print('Mini_GD')
loss, W = Mini_GD(X_train,y_train, lr=0.001)
y_pred = np.dot(X_test,W)
dict_W_Loss[loss] = W
y_hat.append(y_pred)
print()
```
